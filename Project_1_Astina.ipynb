```python
#Replacing NaN values for Normalized Losses with their mean
avg_norm_loss = df["Normalized-Losses"].astype("float").mean(axis=0)
df["Normalized-Losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
#Replacing NaN values for Bore with their mean
avg_bore=df['Bore'].astype('float').mean(axis=0)
df["Bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
#Replacing NaN values for Stroke with their mean
avg_stroke = df["Stroke"].astype("float").mean(axis=0)
df["Stroke"].replace(np.nan, avg_stroke, inplace=True)
```

```python
#Replacing NaN values for Horsepower with their mean
avg_horsepower = df['Horsepower'].astype('float').mean(axis=0)
df['Horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
#Replacing NaN values for Peak-Rpm with their mean
avg_peakrpm=df['Peak-Rpm'].astype('float').mean(axis=0)
df['Peak-Rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
#To find the maximum no. of doors 
df['Num-Of-Doors'].value_counts()
#Replacing NaN values for Num-Of-Doors with their mean i.e. four
df["Num-Of-Doors"].replace(np.nan, "four", inplace=True)
```

```python
#Drop the whole row with NaN in Price column
df.dropna(subset=["Price"], axis=0, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

**Bringing the columns into correct datatypes**

```python
df[["Bore", "Stroke"]] = df[["Bore", "Stroke"]].astype("float")
df[["Normalized-Losses"]] = df[["Normalized-Losses"]].astype("int")
df[["Price"]] = df[["Price"]].astype("float")
df[["Peak-Rpm"]] = df[["Peak-Rpm"]].astype("float")
```

```python
df.dtypes
```

**Bring the original column range to uniform range 0-1**

```python
df['Length'] = df['Length']/df['Length'].max()
df['Width'] = df['Width']/df['Width'].max()
df['Height'] = df['Height']/df['Height'].max()
df[["Height"]].head(10)
```

**In our dataset, "horsepower" is a real valued variable ranging from 48

```python
import matplotlib as plt 
from matplotlib import pyplot
plt.pyplot.hist(df["Horsepower"])
#Plot  hist graph
plt.pyplot.xlabel("Horsepower")
plt.pyplot.ylabel("Count")
plt.pyplot.title("Horsepower Bins")
```

```python
#Building bin arrays
bins = np.linspace(min(df["Horsepower"]), max(df["Horsepower"]), 4)
#Naming bin arrays
group_names = ['Min', 'Mid', 'Max']

df['Horsepower-Binned'] = pd.cut(df['Horsepower'], bins, labels = group_names, include_lowest=True )
df[['Horsepower','Horsepower-Binned']].head(10)
```

**We see the column "fuel-type" has two unique values, "gas" or
"diesel". Regression doesn't understand words, only numbers.
To use this attribute in regression analysis, we convert
"fuel-type" into indicator variables. Also do this for any other
columns?**

```python
df.columns
```

```python
variable_1 = pd.get_dummies(df["Fuel-Type"])
variable_1.head()
```

```python
variable_1.rename(columns={'Fuel-Type-Diesel':'Gas', 'Fuel-Type-Diesel':'Diesel'}, inplace=True)
variable_1.head()
```

```python
#Merge dataframe
df = pd.concat([df, variable_1], axis=1)

# drop original columnFuel-Type from dataframe
df.drop("Fuel-Type", axis = 1, inplace=True)
df.head()
```

```python
#Same for Aspiration column
variable_2 = pd.get_dummies(df["Aspiration"])
variable_2.rename(columns={'Std':'Aspiration-Std', 'Turbo': 'Aspiration-Turbo'}, inplace=True)
variable_2.head()
```

```python
#Merge dataframe
df = pd.concat([df, variable_2], axis=1)

# drop original column Aspiration from dataframe
df.drop("Aspiration", axis = 1, inplace=True)
df.head()
```

**ANOVA: Analysis of Variance**

```python
#Drive-Wheels
grouped_test1=gres1[['Drive-Wheels', 'Price']].groupby(['Drive-Wheels'])
grouped_test1.head(2)
```

```python
gres1
```

```python
grouped_test1.get_group('4wd')['Price']
```

```python
# ANOVA
f_val, p_val = stats.f_oneway(grouped_test1.get_group('fwd')['Price'], grouped_test1.get_group('rwd')['Price'], grouped_test1.get_group('4wd')['Price'])  
 
print( "ANOVA results: F=", f_val, ", P =", p_val)   
```

```python
f_val, p_val = stats.f_oneway(grouped_test1.get_group('fwd')['Price'], grouped_test1.get_group('rwd')['Price'])
 
print( "ANOVA results: F=", f_val, ", P =", p_val)  
```

```python
f_val, p_val = stats.f_oneway(grouped_test1.get_group('4wd')['Price'], grouped_test1.get_group('rwd')['Price'])
 
print( "ANOVA results: F=", f_val, ", P =", p_val)  
```

```python
f_val, p_val = stats.f_oneway(grouped_test1.get_group('4wd')['Price'], grouped_test1.get_group('fwd')['Price'])
 
print( "ANOVA results: F=", f_val, ", P =", p_val)  
```
